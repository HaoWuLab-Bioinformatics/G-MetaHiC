### Process from individual chrom text files

````python

def load_hic(path,hic_resolution):
    hic_dict = {}
    for file in os.listdir(path):
        if file.endswith('txt'):
            print(file)
            # 细胞系没有后缀的
            # chrom = file.split('_')[1]
            # 细胞系有后缀，如gm12878_2
            chrom = file.split('_')[2]
            chrom = chrom.split('.')[0]
            print(chrom)
            hic = pd.read_table(path+file, header = None)
            start = hic[[0,1]].values.min()
            # print(start)
            end = hic[[0,1]].values.max()+hic_resolution
            # print(end)
            hic = hic.pivot(index=0,columns=1,values=2).fillna(0)
            hic = hic.reindex(columns=np.arange(0,end, hic_resolution).astype(int), index=np.arange(0,end, hic_resolution).astype(int),
                      fill_value = 0)
            hic_dict[chrom] = csr_matrix(hic)

    return hic_dict
````

````python
hic_resolution = 1e4

# path = './zvalue/'
path = './gmetahic/datasets/hic/zvalue/CD4_TCells/'
hic_dict = load_hic(path,hic_resolution=hic_resolution)
pickle.dump(hic_dict, open("./gmetahic/datasets/hic/zvalue/CD4_TCells/hic_zscore_dict.p", "wb"), protocol=4)  # save it into a file named save.p
````
